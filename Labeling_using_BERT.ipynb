```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
file_url = f'https://drive.google.com/uc?id=1cIM97LuWSQnnMc9Bh1Y7qDpQs6UsJDEw'


try:
    df = pd.read_csv(file_url, delimiter=',', engine='python', on_bad_lines='skip')
    print(df.head())
except Exception as e:
    print(f"Error: {e}")
```

```python
df.head()
```

```python
from transformers import pipeline

pretrained_name = "w11wo/indonesian-roberta-base-sentiment-classifier"

nlp = pipeline(
    "sentiment-analysis",
    model=pretrained_name,
    tokenizer=pretrained_name,
    truncation=True,
    padding='max_length',
    max_length=512
)
```

```python

df = df.head()  # frac=1 berarti mengambil semua baris dalam urutan acak

# Menginisialisasi daftar untuk prediksi sentimen
sentiment_predictions = []

# Melakukan prediksi sentimen pada semua data
for text_sample in df["reviews"]:
    prediction = nlp(text_sample)
    label = prediction[0]['label']
    score = prediction[0]['score']

    # Mendefinisikan threshold untuk sentimen positif, negatif, dan netral
    if label == 'positive':
        sentiment = 'positive'
    elif label == 'negative':
        sentiment = 'negative'
    else:
        sentiment = 'negative'

    sentiment_predictions.append(sentiment)

# Menambahkan prediksi sentimen ke dalam dataset
df["sentimen"] = sentiment_predictions

# Menyimpan dataset yang telah dimodifikasi ke dalam file CSV baru
new_file_path = 'labeled_sentimen.csv'
df.to_csv(new_file_path, index=False)


```
